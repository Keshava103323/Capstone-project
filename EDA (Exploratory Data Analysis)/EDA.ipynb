```python

# Pandas - Data manipulation
import pandas as pd

# NumPy - Numerical computations
import numpy as np

# Matplotlib - Data visualization
import matplotlib.pyplot as plt

# Scikit-learn - Missing value handling
from sklearn.impute import SimpleImputer
```

```python
# Displays complete information about the dataset
# It shows:
# 1. Number of rows and columns
# 2. Column names
# 3. Data types
# 4. Number of non-null values
# 5. Memory usage

print("Dataset Information")
print("="*50)

df.info()
```

```python
# describe() calculates

# Mean
# Standard Deviation
# Minimum
# Maximum
# Quartiles

print(df.describe())
```

```python

# describe(include='object')

# Displays

# Count

# Unique values

# Most frequent value

# Frequency

print(df.describe(include='object'))
```

```python

# Checks missing values

missing = df.isnull().sum()

print(missing)
```

```python

# Displays number of unique values
# present in every column

for col in df.columns:

    print("Column :", col)

    print("Unique Values :", df[col].nunique())

    print("-"*50)
```

```python
# Displays value counts
# for every categorical column

cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:

    print("\n")

    print("Column :", col)

    print(df[col].value_counts())

    print("="*60)
```

```python
# Histogram shows

# Distribution

# Skewness

# Frequency

num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:

    plt.figure(figsize=(6,4))

    plt.hist(df[col], bins=20)

    plt.title("Histogram of "+col)

    plt.xlabel(col)

    plt.ylabel("Frequency")

    plt.show()
```
