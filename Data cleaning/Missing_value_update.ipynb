```python
# Check missing values in each column
# Algorithm: Missing Value Detection
missing_values = df.isnull().sum()

# Display only columns with missing values
print(missing_values[missing_values > 0])
```

```python
# Step 3: Separate numerical and categorical columns
# Algorithm: Data Type Identification
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
# Print numerical columns
print("Numerical Columns:")
print(list(num_cols))
```

```python
# Print categorical columns
print("\nCategorical Columns:")
print(list(cat_cols))
```

```python
# Step 4: Import SimpleImputer
from sklearn.impute import SimpleImputer
```

```python
# Fill missing numerical values with median
num_imputer = SimpleImputer(strategy='median')
df[num_cols] = num_imputer.fit_transform(df[num_cols])
```

```python
# Fill missing categorical values with mode
cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

```python
# Step 5: Verify missing values
print(df.isnull().sum())
```

```python
df.to_excel("Missing data_removal.xlsx", index=False)
```
